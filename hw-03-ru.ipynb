# Биграммы

С биграммами расшифровка становится не столь тривиальной, поскольку символы в биграммах, построенных на всем тексте и на тестовом подмножестве, могут не совпадать. Для восстановления шифра можем использовать итеративный подход:
1. Обозначим через $R$ исходный текст, через $T$ -- зашифрованный тестовый
2. Для каждого текста посчитаем частотность биграмм в каждом тексте и отсортируем биграммы по частоте появления в порядке убывания.
3. Пока есть нерасшифрованные символы
    1. обозначим:
        1. $X_RY_R$ — биграмма из исходного текста
        2. $X_TY_T$ — биграмма из зашифрованного тестового текста
        3. \# — символ не расшифрован, \_ — символ известен; то есть если $X_TY_T=\#\_$, то имеется в виду, что первый символ неизвестен, а второй - расшифрован 
        4. $D(X_T\rightarrow X_R)$ — известное правило расшифровки
    2. имеем несколько возможностей:
        1. если $X_TY_T=\#\#$, то в списке частотности исходного текста находим биграмму, оба символа которой отсутствуют в $D$, и добавляем в дешифровочный код добавляем пару $X_T\rightarrow X_R, Y_T\rightarrow Y_R$
        2. если $X_TY_T=\#\_$, то в списке частотности исходного текста находим первую биграмму $X_RY_R$, первый символ которой равен $X_T$, а второй символ отсутствует в $D$, и добавляем в дешифровочный код добавляем пару $Y_T\rightarrow Y_R$
        3. если $X_TY_T=\_\#$, то в списке частотности исходного текста находим первую биграмму $X_RY_R$, второй символ которой равен $X_T$, а первый символ отсутствует в $D$, и добавляем в дешифровочный код добавляем пару $X_T\rightarrow X_R$
        4. может случиться так, что тестового примера недостаточно для расшифровки; в таком случае отображение между оставшимися символами назначим случайным образом

```python
def get_decipher_mapping(bigram_counter, encrypted_texts, key_chars=key_chars):
    d, u = dict(), set(key_chars)
    tmp_counter = bigram_counter.copy()
    test_bigram = sum(map(lambda x: Counter(nltk.ngrams(x, 2)), encrypted_texts), start=Counter())
    while u: 
#         print(u)
        for bigram, _ in test_bigram.most_common():
            if bigram[0] in u and bigram[1] in u:
#                 print("Process bigram", *bigram)
                (x, y), _ = tmp_counter.most_common(1)[0]
                del tmp_counter[x, y]
                d[bigram[0]], d[bigram[1]] = x, y
                tmp_counter.pop((y, x), None)
                test_bigram.pop(bigram[::-1], None)
                u -= {*bigram}
#                 print(f"Mappings found and removed: {bigram[0]} -> {x}, {bigram[1]} -> {y}")
                break
            elif bigram[0] in u:
#                 print("Process bigram", *bigram)
                for (x, y), _ in tmp_counter.most_common():
                    if y == d[bigram[1]] and x not in d.values():
                        break
                del tmp_counter[x, y]
                d[bigram[0]] = x
                tmp_counter.pop((y, x), None)
                test_bigram.pop(bigram[::-1], None)
                u -= {*bigram}
#                 print(f"Mapping found and removed: {bigram[0]} -> {x}")
                break
            elif bigram[1] in u:
#                 print("Process bigram", *bigram)
                for (x, y), _ in tmp_counter.most_common():
                    if x == d[bigram[0]] and y not in d.values():
                        break
                del tmp_counter[x, y]
                d[bigram[1]] = y
                tmp_counter.pop((y, x), None)
                test_bigram.pop(bigram[::-1], None)
                u -= {*bigram}
#                 print(f"Mapping found and removed: {bigram[1]} -> {y}")
                break
        else:
            unmatched = set(key_chars) - set(d.values())
            for x, y in zip(u, unmatched):
                d[x] = y
            u.clear()
    return d
```

# MCMC

Для начала нам необходима функция, которая бы каждому тексту ставила в соответствие действительное число. В качестве таковой можно использовать следующую: $$S(X)=\prod_{\alpha\beta\in X}f_R(\alpha, \beta)^{f_X(\alpha, \beta)},$$ где $f_T(\alpha, \beta)$ — частота биграммы $\alpha\beta$ в тексте $T$,

$R$ — исходный текст,

$X$ — текст для оценки.

Далее нам нужна функция, исходя из значений которой мы будем решать, будем ли мы принимать новое изменение в ходе МСМС. В качестве таковой возьмем следующую. $$A(X,Y)=\min\left(1,\frac{S(X)}{S(Y)}\right)$$

Для удобства, а также для того чтобы избавиться от возможных ошибок при работе с малыми числами, будем пользоваться логарифмом этой функции. $$L(X,Y)=\ln A(x,y)=\min\left(0, \ln\frac{S(X)}{S(Y)}\right)=\min\left(0, \ln S(X) - \ln S(Y)\right)=$$
$$=\min\left(0, \ln{\prod_{\alpha\beta\in X}f_R(\alpha, \beta)^{f_X(\alpha, \beta)}} - \ln{\prod_{\alpha\beta\in Y}f_R(\alpha, \beta)^{f_Y(\alpha, \beta)}}\right)=\min\left(0, \sum_{\alpha\beta\in X}f_X(\alpha, \beta)\ln{f_R(\alpha, \beta)}-\sum_{\alpha\beta\in Y}f_Y(\alpha, \beta)\ln{f_R(\alpha, \beta)}\right)$$

```python
def score(text, reference_freqs=bigram_counter):
    f_text = Counter(nltk.ngrams(text, 2))
    s = 0
    for x, y in f_text:
        s += f_text[x, y] * np.log(reference_freqs.get((x, y), 1)) # на случай, если вдруг окажется, что нет биграммы в исходном тексте, хотя в течение валидации это и невозможно
    return s
```

```python
def get_next_mapping(cur_mapping):
    mapping_lst_ = list(cur_mapping)
    key1 = np.random.choice(mapping_lst_)
    key2 = np.random.choice(mapping_lst_)
    if key1 == key2:  # если вдруг мы выбрали один и тот же символ
        return get_next_mapping(cur_mapping)
    new_mapping = cur_mapping.copy()
    new_mapping[key1], new_mapping[key2] = cur_mapping[key2], cur_mapping[key1]
    return new_mapping
```

```python
n = 100
accuracy = []
for _ in range(n):
    # случайный шифр
    encr_mapping = dict(zip(key_chars, np.random.permutation(list(key_chars))))
    # случайная тестовая выборка
    test_samples = np.random.choice(ru_sentences, size=10, replace=False)
    # зашифрованный текст
    encrypted_texts = [apply_encryption(test_, encr_mapping) for test_ in test_samples]


    best_decipher = None
    best_score = None
    N_ITER = 10000
    cur_mapping = dict(zip(key_chars, np.random.permutation(list(key_chars))))
    for i in range(1, N_ITER + 1):
        next_mapping = get_next_mapping(cur_mapping)
        decrypted_current = [apply_encryption(test_, cur_mapping) for test_ in encrypted_texts]
        decrypted_next = [apply_encryption(test_, next_mapping) for test_ in encrypted_texts]
        current_score = score(''.join(decrypted_current))
        next_score = score(''.join(decrypted_next))
        if best_score is None or next_score > best_score:
            best_score = next_score
            best_decipher = next_mapping
        elif np.random.uniform() <= min(1, np.exp(next_score - current_score)):
            cur_mapping = next_mapping.copy()
    decrypted_text = ''.join([apply_encryption(test_, best_decipher) for test_ in encrypted_texts])
    real_text = ''.join(test_samples)
    if n < 10:
        print(real_text[:150])
        print(decrypted_text[:150])
        print()
    accuracy.append(sum(x == y for x, y in zip(decrypted_text, real_text)) / len(decrypted_text))
print(sum(accuracy) / n)    
```

# Триграммы

Переход к триграммам производится достаточно просто: достаточно адаптировать функцию оценки.

$$S(X)=\prod_{\alpha\beta\gamma\in X}f_R(\alpha, \beta, \gamma)^{f_X(\alpha, \beta, \gamma)},$$ где $f_T(\alpha, \beta, \gamma)$ — частота триграммы $\alpha\beta\gamma$ в тексте $T$,

$R$ — исходный текст,

$X$ — текст для оценки.

$$A(X,Y)=\min\left(1,\frac{S(X)}{S(Y)}\right)$$

$$L(X,Y)=\ln A(x,y)=\min\left(0, \ln\frac{S(X)}{S(Y)}\right)=\min\left(0, \ln S(X) - \ln S(Y)\right)=$$
$$=\min\left(0, \ln{\prod_{\alpha\beta\gamma\in X}f_R(\alpha, \beta, \gamma)^{f_X(\alpha, \beta, \gamma)}} - \ln{\prod_{\alpha\beta\gamma\in Y}f_R(\alpha, \beta, \gamma)^{f_Y(\alpha, \beta, \gamma)}}\right)=\min\left(0, \sum_{\alpha\beta\gamma\in X}f_X(\alpha, \beta, \gamma)\ln{f_R(\alpha, \beta, \gamma)}-\sum_{\alpha\beta\gamma\in Y}f_Y(\alpha, \beta, \gamma)\ln{f_R(\alpha, \beta, \gamma)}\right)$$

```python
trigram_counter = sum(map(lambda x: Counter(nltk.ngrams(x, 3)), ru_sentences), start=Counter())
```

```python
def score3(text, reference_freqs=trigram_counter):
    f_text = Counter(nltk.ngrams(text, 3))
    s = 0
    for x, y, z in f_text:
        s += f_text[x, y, z] * np.log(reference_freqs.get((x, y, z), 1)) # на случай, если вдруг окажется, что нет биграммы в исходном тексте, хотя в течение валидации это и невозможно
    return s
```
